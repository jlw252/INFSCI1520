```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import string
from collections import Counter
import numpy as np

df = pd.read_csv('Taylor_Swift_Albums.csv', encoding='ISO-8859-1')
df['Album'] = df['Album'].str.replace(" \(Taylor's Version\)", '', regex=False)
df = df.dropna(subset=['Lyrics'])

album_word_counts = {}
word_album_map = {}


for _, row in df.iterrows():
    album = row['Album']
    lyrics = row['Lyrics'].lower()
    lyrics = lyrics.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    words = lyrics.split()

    if album not in album_word_counts:
        album_word_counts[album] = Counter()
    album_word_counts[album].update(words)

    for word in set(words):
        if word not in word_album_map:
            word_album_map[word] = Counter()
        word_album_map[word][album] += 1

total_counts = sum(album_word_counts.values(), Counter())

#identify which album each word is most associated with
word_color_map = {word: word_album_map[word].most_common(1)[0][0] for word in total_counts}

#colors
colors = {
    "Taylor Swift": "#98FF98",   # Mint green
    "Fearless": "#FFD700",       # Gold
    "Speak Now": "#A020F0",      # Purple
    "Red": "#FF0000",            # Red
    "1989": "#40E0D0",           # Turquoise
    "Reputation": "#000000",     # Black
    "Lover": "#FF69B4",          # Pink
    "Folklore": "#B4A89C",       # Beige grey
    "Evermore": "#FFFDD0",       # Cream
    "Midnights": "#191970",      # Navy
    "TTPD": "#909090"            # Dark grey
}

#define color function for word cloud
def updated_color_func(word, *args, **kwargs):
    album = word_color_map.get(word, "Taylor Swift")
    hex_color = colors.get(album, "#000000")
    return hex_color

wordcloud = WordCloud(
    width=1600,
    height=800,
    background_color='white',
    color_func=updated_color_func
).generate_from_frequencies(total_counts)


plt.figure(figsize=(20, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
#plt.title('Taylor Swift Lyrics: Most Common Words by Album Color', fontsize=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Taylor_Swift_Albums.csv', encoding='ISO-8859-1')

df['Album'] = df['Album'].str.strip().str.title()
df['Genre'] = df['Genre'].astype(str).str.lower().str.strip()
df['Genre List'] = df['Genre'].str.split(',')

df_exploded = df.explode('Genre List')
df_exploded['Genre List'] = df_exploded['Genre List'].str.strip()

def map_genre_to_broad(genre):
    genre = genre.strip().lower()
    if 'pop' in genre and genre != 'country pop':
        return 'Pop'
    elif genre in ['country', 'country pop', 'country rock']:
        return 'Country'
    elif genre in ['folk', 'folk pop', 'indie folk', 'singer-songwriter', 'acoustic']:
        return 'Folk'
    elif genre in ['rock', 'alternative rock', 'post-grunge', 'soft rock', 'alternative', 'new wave']:
        return 'Rock'
    else:
        return 'Other'

df_exploded['Refined Genre'] = df_exploded['Genre List'].apply(map_genre_to_broad)

refined_genre_album_counts = df_exploded.groupby(['Album', 'Refined Genre']).size().reset_index(name='Count')
release_order = [
    "Taylor Swift", "Fearless", "Speak Now", "Red", "1989", "Reputation",
    "Lover", "Folklore", "Evermore", "Midnights", "Ttpd"
]
refined_genre_album_counts['Album'] = pd.Categorical(refined_genre_album_counts['Album'], categories=release_order, ordered=True)
refined_genre_album_counts = refined_genre_album_counts.sort_values(['Album', 'Count'], ascending=[True, False])

#colors
color= {
    'Pop': '#FFB6C1',       # Light pink
    'Country': '#ffca7d',   # Pastel Orange
    'Folk': '#90EE90',      # Light green
    'Rock': '#87CEFA',      # Light sky blue
    'Other': '#D8BFD8'      # Thistle
}

plt.figure(figsize=(14, 8))
sns.barplot(
    data=refined_genre_album_counts,
    x='Album',
    y='Count',
    hue='Refined Genre',
    palette=color
)
plt.xlabel('Album', fontsize=14)
plt.ylabel('Number of Songs', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.legend(
    title='Genre',
    loc='upper left',
    bbox_to_anchor=(0.01, 0.98),
    prop={'size': 12},
    title_fontsize=13
)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

df = pd.read_csv("eras tour dates.csv", encoding="ISO-8859-1")
df["Full Location"] = df["City"] + ", " + df["country"]
geolocator = Nominatim(user_agent="eras_tour_mapper", timeout=10)
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

def safe_geocode(location):
    try:
        return geocode(location)
    except:
        return None

#print("Geocoding locations... This may take a few minutes.")
df["location"] = df["Full Location"].apply(safe_geocode)
df["point"] = df["location"].apply(lambda loc: (loc.latitude, loc.longitude) if loc else (None, None))
df = df.dropna(subset=["point"])

gdf = gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(
        [pt[1] for pt in df["point"]],
        [pt[0] for pt in df["point"]]
    ),
    crs="EPSG:4326"
)

world = gpd.read_file("ne_110m_admin_0_countries.shp")
fig, ax = plt.subplots(figsize=(15, 10))
world.plot(ax=ax, color="#f4f4f4", edgecolor="#cccccc", linewidth=0.3)
gdf.plot(ax=ax, color="#e60073", markersize=100, alpha=0.9)
plt.title("Taylor Swift's Eras Tour", fontsize=20, fontweight='bold', color="#e60073")

plt.axis("off")
plt.tight_layout()

plt.savefig("eras_tour_map_clean_dots.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.colors import LinearSegmentedColormap, Normalize
import warnings

warnings.filterwarnings("ignore", message="Glyph.*missing from current font")

df = pd.read_csv("Taylor_Swift_Streams.csv", encoding="ISO-8859-1")
df["Streams"] = df["Streams"].str.replace(",", "").astype("int64")
df["Daily"] = df["Daily"].str.replace(",", "").astype(float)

top_songs = df.sort_values(by="Streams", ascending=False).head(15)
top_songs_sorted = top_songs.sort_values(by="Streams", ascending=True).reset_index(drop=True)

gradient = LinearSegmentedColormap.from_list(
    "gradient",
    [
        "#301934",  # deep purple
        "#8B008B",  # dark magenta
        "#FF0000",  # red
        "#FF5A9F",  # darker hot pink
        "#FFB6C1",  # light pink
        "#FFD1DC",  # pale pink
        "#FFE4E1",  # misty rose
        "#FFF0F5"   # lavender blush
    ],
    N=256
)

norm_total = Normalize(vmin=top_songs_sorted["Streams"].min(), vmax=top_songs_sorted["Streams"].max())
norm_daily = Normalize(vmin=top_songs_sorted["Daily"].min(), vmax=top_songs_sorted["Daily"].max())

total_colors = [gradient(norm_total(val)) for val in top_songs_sorted["Streams"]]
daily_colors = [gradient(norm_daily(val)) for val in top_songs_sorted["Daily"]]

plt.figure(figsize=(14, 10))
sns.set(style="whitegrid")

#Total Streams (left)
for i, (title, val) in enumerate(zip(top_songs_sorted["Song Title"], top_songs_sorted["Streams"])):
    plt.barh(
        y=i,
        width=-val,
        color=total_colors[i],
        edgecolor="none"
    )

#Daily Streams (right, scaled)
scaling_factor = top_songs_sorted["Streams"].max() / top_songs_sorted["Daily"].max() * 0.6
for i, (title, val) in enumerate(zip(top_songs_sorted["Song Title"], top_songs_sorted["Daily"])):
    plt.barh(
        y=i,
        width=val * scaling_factor,
        color=daily_colors[i],
        edgecolor="none"
    )

for i, (total, daily) in enumerate(zip(top_songs_sorted["Streams"], top_songs_sorted["Daily"])):
    plt.text(-total - 1e7, i, f'{int(total/1e6)}M', va='center', ha='right', fontsize=9, color='black')
    plt.text(daily * scaling_factor + 1e7, i, f'{int(daily/1e3)}K', va='center', ha='left', fontsize=9, color='black')

plt.yticks(ticks=range(len(top_songs_sorted)), labels=top_songs_sorted["Song Title"], fontsize=11)
plt.axvline(0, color="black", linewidth=1)
plt.title("Taylor Swift: Total vs Daily Streams", fontsize=16, fontweight="bold")
plt.xlabel("Streams ( Total in Millions | Daily in Thousands )", fontsize=12)
plt.xticks(
    ticks=np.arange(-3e9, 3e9 + 1, 5e8),
    fontsize=10
)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

df = pd.read_csv("Taylor_Swift_Streams.csv", encoding="ISO-8859-1")
df["Streams"] = df["Streams"].str.replace(",", "").astype("int64")
df["Daily"] = df["Daily"].str.replace(",", "").astype(float)

#Top 12 songs by daily streams (descending)
top_12 = df.sort_values(by="Daily", ascending=False).head(12).reset_index(drop=True)

songs = top_12["Song Title"][::-1]
total_streams = top_12["Streams"][::-1]
daily_streams = top_12["Daily"][::-1]
y_pos = range(len(songs))

#create gradient
gradient = LinearSegmentedColormap.from_list(
    "gradient",
    [
        "#301934", "#8B008B", "#FF0000", "#FF5A9F",
        "#FFB6C1", "#FFD1DC", "#FFE4E1", "#FFF0F5"
    ]
)

#scale daily streams for visual balance
scaling_factor = total_streams.max() / daily_streams.max()
scaled_daily_streams = daily_streams * scaling_factor

total_ranks = total_streams.rank(ascending=True).astype(int) - 1  # 0 (lowest) to 11 (highest)
daily_ranks = daily_streams.rank(ascending=True).astype(int) - 1

left_colors = [gradient(rank / 11) for rank in total_ranks]
right_colors = [gradient(rank / 11) for rank in daily_ranks]

fig, ax = plt.subplots(figsize=(14, 8))

ax.barh(y_pos, [-v for v in total_streams], color=left_colors, label="Total Streams")
ax.barh(y_pos, scaled_daily_streams, color=right_colors, label="Daily Streams (Scaled)")

for i, (t, d, sd) in enumerate(zip(total_streams, daily_streams, scaled_daily_streams)):
    ax.text(-t - 2e7, i, f'{t/1e6:.0f}M', va='center', ha='right', fontsize=9, color='#333')
    ax.text(sd + 2e7, i, f'{d/1e3:.0f}K', va='center', ha='left', fontsize=9, color='#333')
    
ax.set_yticks(y_pos)
ax.set_yticklabels(songs)
ax.set_xlabel("Streams (Daily Scaled to Match Total)")
ax.axvline(0, color='black', linewidth=1)
ax.set_title("Top Daily Songs", pad=20)
ax.legend(loc="lower right")

plt.tight_layout()
plt.show()
```
